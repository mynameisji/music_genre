### <**데이터셋 요소 설명>**

아래 모든 요소는 높은 값일수록 해당 요소의 경향이 강하다.

**bpm** : 음악의 속도를 숫자로 표시. 숫자가 클수록 빠르다.

**nrgy(energy)** : 에너지가 클수록 빠르고 크고 시끄러운 음악이다. ex) 헤비메탈은 에너지가 높고, 모차르트의 자장가는 에너지가 낮다.

**dnce(danceability)** : 춤추기에 적합한 음악인지 판단. 템포, 리듬 안정성, 비트 강도 등이 전체적으로 일정한지 조합하여 정해진다.

**dB(loudness)** : 값이 클수록 소리가 큰 음악이다.

**live(liveness)** : 값이 클수록 (스튜디오 녹음이 아닌) 실황 녹음일 확률이 높다.

**val(valence)** : 밝고 긍정적인 느낌의 정도

**dur(duration)** : 차트에 올라와있던 기간

**acous(acoustic)** : 노래가 어쿠스틱한 정도 (어쿠스틱한 소리는 전자음이 섞이지 않은 생악기 소리)

**spch(speechiness)** : 값이 클수록 노래에 말이 많이 나온다. 트랙에 음성이 포함되어 있는지 여부를 측정. 

**pop(popularity)** : 인기 정도.

#**2. 2010년대는 rock의 암흑기, elctronic의 부흥기, 그리고 대세는 hip hop**

- 1990년대, 2000년대, 2010년대 인기있었던 대중음악 장르 중 대표적으로 rock/metal, hip hop, electronic music의 변천사를 확인해보고자 한다.

- 'rock'과 'metal'을 키워드로 추출하여 rock 데이터프레임 생성 -> 'top genre'값을 일괄적으로 rock으로 입력

- 'hip hop'과 'rap'을 키워드로 추출하여 hiphop 데이터프레임 생성 -> 'top genre'값을 일괄적으로 'hip hop'으로 입력

- 'electro'와 'big room'을 키워드로 추출하여 electronic 데이터프레임 생성 -> 'top genre'값을 일괄적으로 'electronic'으로 입력

위 세개의 데이터프레임을 다시 또 합쳐서 1990s~2010s 동안 rock, hip hop, electronic 음악 장르의 선호도 변천사를 시각화하려고 한다.

1) 1990s ~ 2010s 중 rock 장르 데이터프레임 만들기

```
df_rock = df[df['top genre'].str.contains('rock')]
print(df_rock.shape)
df_rock.head()
```

```
df_metal = df[df['top genre'].str.contains('metal')]
print(df_metal.shape)
df_metal.head()
```

```
rock = pd.concat([df_rock, df_metal])
rock = rock.sort_values(by=['year'], axis=0, ascending = True)
rock = rock.reset_index(drop=True)
rock.loc[0:17, ['top genre']] = 'rock/metal'
print(rock.shape)
rock.head()
```

2) 1990s ~ 2010s 중 hip hop 장르 데이터프레임 만들기

```
df_hiphop = df[df['top genre'].str.contains('hip hop')]
print(df_hiphop.shape)
df_hiphop.head()
```

```
df_rap = df[df['top genre'].str.contains('rap')]
print(df_rap.shape)
df_rap.head()
```

```
hiphop = pd.concat([df_hiphop, df_rap])
hiphop = hiphop.sort_values(by=['year'], axis=0, ascending = True)
hiphop = hiphop.reset_index(drop=True)
hiphop.loc[0:36, ['top genre']] = 'hiphop'
print(hiphop.shape)
hiphop.head()
```

3) 1990s ~ 2010s 중 electronic 장르 데이터프레임 만들기

```
df_electro = df[df['top genre'].str.contains('electro')]
print(df_electro.shape)
df_electro.head()
```

```
df_bigroom = df[df['top genre'].str.contains('big room')]
print(df_bigroom.shape)
df_bigroom.head()
```

```
electronic = pd.concat([df_electro, df_bigroom])
electronic = electronic.sort_values(by=['year'], axis=0, ascending = True)
electronic = electronic.reset_index(drop=True)
electronic.loc[0:19, ['top genre']] = 'electronic'
print(electronic.shape)
electronic.head()
```
